<h5> DEFINICION DE CLUSTERING JERARQUICO Y PROCEDIMIENTO </h5>

Metodo aglomerativo donde cada punto empieza  siendo un unico cluster  y en cada paso los vamos juntando hasta obtener un unico cluster basado en las similitude de entre observaciones

1. Empezamos con cada observacion enun cluster (tenemos N cluster)
2. Encontramos la menor distancia en la matrices de distancias y unimos las dos observaciones en un nuevo cluster
3. Recalculamos las distancias entre los clusters antiguos y el nuevo recien creado con alguno de los metodos de enlace

METODOS DE ENLACE
#(enlace simple o distancia minima entre cualquiera de los puntos del cluster)
Enlace compelto, la distancia entre los dos clusters es el maximo de las distancias entre cualqiier dos puntos del cluster 1 y 2
enlace promedio(promedio de las distancias entre dos puntos del cluster)
enlace centroide (punto medio del cluster 1 y 2)
Enlace de War
UN cluster que minimiza el incremento del error combinando la suma de los cuadrados de la varianza se a un cluster ya existente para formar una nuevo.

se repite el paso 2 y 3 hasta que solo queda un cluster con Las N observaciones

# Clustering jerárquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage # dendograma para hacer representacion de jerarquica y linkage para hacer los enlaces
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward")#calcula el enlace que le metamos
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)#z la matriz, parametro que rota las etiquetas, tamaño fuente para que sea mas pequeño
plt.show()
```

```python
Z = linkage(data[movies], "average") #enlace promedio
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "complete")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```
