* Importar el primer fichero
* Hacemos un bucle para ir recorriendo todos y cada uno de los ficheros. 
    * Importante tener una consistencia en el nombre de los ficheros 
    * Importamos los ficheros uno a uno
    * Cada uno de ellos debe apendizarse (aadirse al final) del primer fichero que ya habamos cargado
* Repetimos el bucle hasta que no queden ficheros

```python
#filepath = "/content/drive/My Drive/Curso Machine Learning con Python/datasets/distributed-data/"

filepath="/home/juan/Documentos/DATA SCIENCE/python-ml-course-master/datasets/distributed-data/"
filename="001.csv"

data = pd.read_csv(filepath+"001.csv")
final_length = len(data)

for i in range(2,333):
    if i < 10:
        filename = "00" + str(i)
    if 10 <= i < 100:
        filename = "0" + str(i)
    if i >= 100:
        filename = str(i)
    file = filepath + filename + ".csv"
    
    temp_data = pd.read_csv(file)
    final_length += len(temp_data)
    
    data = pd.concat([data, temp_data], axis = 0)
```

## Inner Join
* Devuelve un data frame con las filas que tienen valor tanto en el primero como en el segundo data frame que estamos uniendo
* El nmero de filas ser igual al nmero de filas **comunes** que tengas ambos data sets
    * Data Set A tiene 60 filas
    * Data Set B tiene 50 filas
    * Ambos comparten 30 filas
    * Entonces A Inner Join B tendr 30 filas
* En trminos de teora de conjuntos, se trata de la interseccin de los dos conjuntos

```python
Image(filename="resources/inner-join.png")
```

```python
# data_main contiene toda la info
# data_country_dlt le falta la info de 7 atletas
merged_inner = pd.merge(left = data_main, right = data_country_dlt,
                       how = "inner", left_on = "Athlete", right_on = "Athlete")
```

```python
len(merged_inner)
```

```python
merged_inner.head()
```

## Left Join
* Devuelve un data frame con las filas que tuvieran valor en el dataset de la izquierda, sin importar si tienen correspondencia en el de la derecha o no.
* Las filas del data frame final que no correspondan a ninguna fila del data frame derecho, tendrn NAs en las columnas del data frame derecho.
* El nmero de filas ser igual al nmero de filas del data frame izquierdo
    * Data Set A tiene 60 filas
    * Data Set B tiene 50 filas
    * Entonces A Left Join B tendr 60 filas
* En trminos de teora de conjuntos, se trata del propio data set de la izquierda quien, adems tiene la interseccin en su interior.

```python
Image(filename="resources/left-join.png")
```

```python
merged_left = pd.merge(left = data_main, right = data_country_dlt, 
                      how = "left", left_on = "Athlete", right_on = "Athlete")
len(merged_left)
```

```python
merged_left.head()
```

## Right Join
* Devuelve un data frame con las filas que tuvieran valor en el dataset de la derecha, sin importar si tienen correspondencia en el de la izquierda o no.
* Las filas del data frame final que no correspondan a ninguna fila del data frame izquierdo, tendrn NAs en las columnas del data frame izquierdo.
* El nmero de filas ser igual al nmero de filas del data frame derecho
    * Data Set A tiene 60 filas
    * Data Set B tiene 50 filas
    * Entonces A Right Join B tendr 50 filas
* En trminos de teora de conjuntos, se trata del propio data set de la derecha quien, adems tiene la interseccin en su interior.

```python
Image(filename="resources/right-join.png")
```

```python
merged_right = pd.merge(left = data_main_dlt, right = data_country_dp,
                       how = "right", left_on = "Athlete", right_on = "Athlete")
len(merged_right)
```

## Outer Join
* Devuelve un data frame con todas las filas de ambos, reemplazando las ausencias de uno o de otro con NAs en la regin especfica..
* Las filas del data frame final que no correspondan a ninguna fila del data frame derecho (o izquierdo), tendrn NAs en las columnas del data frame derecho (o izquierdo).
* El nmero de filas ser igual al mximo nmero de filas de ambos data frames
    * Data Set A tiene 60 filas
    * Data Set B tiene 50 filas
    * Ambos comparten 30 filas
    * Entonces A Outer Join B tendr 60 + 50 - 30 = 80 filas
* En trminos de teora de conjuntos, se trata de la unin de conjuntos.

```python
Image(filename="resources/outer-join.png")
```

```python
data_country_jb = data_country_dlt.append(
    {
        "Athlete": "Juan Gabriel Gomila",
        "Country": "Espaa"
    },ignore_index = True
)
```
